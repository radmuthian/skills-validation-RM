```python
%%bigquery
CREATE SCHEMA IF NOT EXISTS weather_dataset
OPTIONS(location="us");
```

```python
%%bigquery
LOAD DATA OVERWRITE weather_dataset.raw_weather
FROM FILES (
  format = 'CSV',
  uris = ['gs://labs.roitraining.com/data-to-ai-workshop/weather_data.csv'],
  skip_leading_rows = 1
);
```

```python
%%bigquery
SELECT * FROM weather_dataset.raw_weather LIMIT 5;
```

```python
%%bigquery
CREATE OR REPLACE MODEL weather_dataset.gemini_model
REMOTE WITH CONNECTION `us.gemini_conn`
OPTIONS (ENDPOINT = 'gemini-1.5-flash');
```

```python
%%bigquery
CREATE OR REPLACE TABLE weather_dataset.weather_with_reports AS
SELECT
  *,
  ml_generate_text_result AS weather_report
FROM
  ML.GENERATE_TEXT(
    MODEL `weather_dataset.gemini_model`,
    (
      SELECT
        *,
        CONCAT(
          'Based on a temperature of ', temperature,
          ' degrees, wind speed of ', wind_speed,
          ' mph, and humidity of ', humidity,
          '%, write a short, professional weather warning or report.'
        ) AS prompt
      FROM
        `weather_dataset.raw_weather`
    ),
    STRUCT(
      0.2 AS temperature,
      250 AS max_output_tokens
    )
  );
```
